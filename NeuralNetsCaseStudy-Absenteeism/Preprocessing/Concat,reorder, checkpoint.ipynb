```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('Absenteeism-data.csv')
```

```python
df = raw_csv_data.copy()
```

```python
df
```

```python
df_no_age = df.drop(['Age'],axis=1)
```

```python
df_no_age
```

```python
age_dummies = pd.get_dummies(df['Age'],drop_first=True)
```

```python
age_dummies
```

```python
age_dummies.iloc[:,11:]
```

```python
age_younger = age_dummies.iloc[:,:11].max(axis=1)
age_older = age_dummies.iloc[:,11:].max(axis=1)
```

```python
df_concatenated = pd.concat([df_no_age, age_younger, age_older], axis=1)
```

```python
df_concatenated
```

```python
df_concatenated.columns.values
```

```python
df_concatenated.columns = ['ID', 'Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Daily Work Load Average', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours', 'Age_younger', 'Age_older']
```
